```python
from transformers import pipeline

# Sentiment analysis
classifier = pipeline("sentiment-analysis")
classifier(["The capital city of ethiopia is not mekelle","The minucipality of ethipia and the whole globe is machine learning and AI"])
```

```python
#Zero - shot Classification
zero_shot = pipeline("zero-shot-classification")
zero_shot("This is an introduction to science and Technology", candidate_labels = ["Education","Health","Agriculture"])
```
